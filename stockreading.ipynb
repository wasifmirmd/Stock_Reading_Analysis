```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
```

```python
start = datetime.datetime(2017,7,11)
end = datetime.datetime(2022,7,11)
```

```python
stock = web.DataReader("AAPL",'yahoo',start, end)
```

```python
stock.head()
```

```python
stock['Open'].plot(label = 'AAPL Open price',figsize=(40,20))
stock['Close'].plot(label = 'AAPL Open price')
stock['High'].plot(label = 'AAPL Open price')
stock['Low'].plot(label = 'AAPL Open price')
plt.legend()
plt.title('Apple Stock Prices')
plt.ylabel('Stock Price')
plt.show()
```

```python
stock['Volume'].plot(figsize=(17,5))
plt.title('Volume traded by Apple')
```

```python
stock.to_csv('apple.csv')
```
